```python
import pandas as pd
import tensorflow as tf
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder , StandardScaler 
```

```python
cdf = pd.read_csv("")
cdf
```

# Preprocessing

```python
# Cheking the columns with object type
df = cdf
df.dtypes
```

```python
#Removing Car_id column
df = df.drop(['car_ID' , 'CarName'] , axis=1)
```

```python
object_columns = []

#find the object datas
for col in df.columns:
    if df[col].dtype == 'object':
        object_columns.append(col)

print(object_columns)
```

```python
nonobject_columns = []

# find the non object datas
for col in df.columns:
    if df[col].dtype != 'object':
        nonobject_columns.append(col)


print(nonobject_columns)
```

```python
# Drop rows with missing values
df = df.dropna()
# Remove duplictes
df = df.drop_duplicates()
```

```python
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col]= le.fit_transform(df[col])
df.head()
```

```python
nonobject_columns.remove('price')
print(nonobject_columns)
```

```python
X = df.drop(['price'], axis=1)
y = df['price']
# Split data to X_train and y_train
X_train, X_test, y_train, y_test = train_test_split(X,y ,test_size = 0.15 ,random_state = 10)
X_train0 , X_val , y_train0 , y_val = train_test_split(X_train,y_train ,test_size = 0.15 ,random_state = 10 )
```

```python
scaler = StandardScaler()
X_train0[nonobject_columns]=scaler.fit_transform(X_train0[nonobject_columns])
X_val[nonobject_columns]= scaler.transform(X_val[nonobject_columns])
X_test[nonobject_columns] = scaler.transform(X_test[nonobject_columns])
```

```python
# Create a StandardScaler object
scaler2 = StandardScaler()
X_train0[object_columns]=scaler2.fit_transform(X_train0[object_columns])
X_val[object_columns] = scaler2.transform(X_val[object_columns])
X_test[object_columns] = scaler2.transform(X_test[object_columns])
```

```python
batch_size=10
train_data = tf.data.Dataset.from_tensor_slices((X_train0, y_train0))
train_data = train_data.shuffle(buffer_size=len(y_train0)).batch(batch_size)
val_data = tf.data.Dataset.from_tensor_slices((X_val, y_val))
val_data = val_data.batch(batch_size)
```

# Define a model

```python
# Define the neural network model
model = keras.models.Sequential([
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(1)])
```

```python
loss_fn = tf.keras.losses.MeanSquaredError()
```

```python
optimizer = tf.keras.optimizers.Adam()
train_acc_metric = tf.keras.metrics.MeanAbsoluteError()
val_acc_metric = tf.keras.metrics.MeanAbsoluteError()
```

```python
num_epochs = 101
train_losses = []
train_accuracies = []
val_losses = []
val_accuracies = []
for epoch in range(num_epochs):
    loss_train = []
    loss_vall = []
    for step, (x_batch_train, y_batch_train) in enumerate(train_data):
        with tf.GradientTape() as g:
            logits = model(x_batch_train , training=True)
            loss_value = loss_fn(y_true=y_batch_train , y_pred = logits)
        gradients = g.gradient(loss_value, model.trainable_weights)
        optimizer.apply_gradients(zip(gradients, model.trainable_weights))
        train_acc_metric.update_state(y_batch_train, logits)
        loss_train.append(loss_value)
    for step, (x_batch_val, y_batch_val) in enumerate(val_data):
        logits_val = model(x_batch_val)
        loss_value_val = loss_fn(y_true=y_batch_val, y_pred=logits_val)
        val_acc_metric.update_state(y_batch_val, logits_val)
        loss_vall.append(loss_value_val)
    train_acc = train_acc_metric.result()
    val_acc = val_acc_metric.result()
    train_accuracies.append(train_acc)
    val_accuracies.append(val_acc)
    losses_train_mean = np.mean(loss_train)
    losses_val_mean = np.mean(loss_vall)
    train_losses.append(losses_train_mean)
    val_losses.append(losses_val_mean)
    if epoch == 1 or epoch % 10 == 0 and epoch != 0:
        print('\n Epoch %s: Train loss: %.4f  Validation Loss: %.4f, Train Accuracy: %.4f, Validation Accuracy %.4f' % (epoch, float(losses_train_mean), float(losses_val_mean), float(train_acc), float(val_acc)))

    train_acc_metric.reset_states()
    val_acc_metric.reset_states()
```
